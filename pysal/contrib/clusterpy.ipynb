```python
def addArray2Layer(array, layer, names=None):
    n,k = array.shape
    if not names:
        names = ["X_%d"% v for v in range(k)]
        
    for j,name in enumerate(names):
        v = {}
        for i in xrange(n):
            v[i] = array[i,j]
        layer.addVariable([name], v)

    
```

```python
addArray2Layer(pci, usy)
```
